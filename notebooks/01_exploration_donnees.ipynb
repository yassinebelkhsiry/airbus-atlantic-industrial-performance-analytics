# 01 - Exploration des données industrielles

Objectif : comprendre la structure du dataset, évaluer la qualité des données et identifier les premiers signaux de performance industrielle.

```python
from pathlib import Path
import pandas as pd

ROOT_DIR = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
DATA_PATH = ROOT_DIR / 'data' / 'raw' / 'donnees_production_airbus_atlantic.csv'

df = pd.read_csv(DATA_PATH, parse_dates=['date_production'], encoding='utf-8')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all').T
```

```python
qualite = df.isna().sum().to_frame('valeurs_manquantes')
qualite['taux_manquant'] = qualite['valeurs_manquantes'] / len(df)
qualite.sort_values('valeurs_manquantes', ascending=False)
```

```python
df.groupby(['site', 'ligne_production'])['quantite_produite'].sum().sort_values(ascending=False).head(10)
```

## Premières observations

- Les données couvrent plusieurs sites, lignes et types de pièces.
- Les variables de qualité permettent de calculer les taux de conformité et de non-conformité.
- Les variables délai, coût et signaux machine permettent d'analyser les causes possibles d'écart de performance.
